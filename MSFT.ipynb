## MODEL  (GRU)

```python
training_set=df2['Open']
training_set=pd.DataFrame(training_set)
training_set
```

```python
# Feature Scaling
import pickle
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0, 1))

training_set_scaled = scaler.fit_transform(training_set)
```

```python
training_size = df2.shape[0]*0.80
testing_size = df.shape[0] - training_size
print("Training Size:"+str(training_size))
print("Testing Size:"+str(testing_size))
```

```python
##Creating Training Set with step of 60-minutes
```

```python
import numpy as np
X_train = []
y_train = []
for i in range(60, 8844):
    X_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)

# Reshaping
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1],1))
```

```python
X_train.shape
```

```python
## Creating Testing Dataset with 60 time stamps
```

```python
import numpy as np
X_test = []
y_test = []
date=[]
for i in range(60, 8805):
    X_test.append(training_set_scaled[i-60:i, 0])
    y_test.append(training_set_scaled[i, 0])
    date.append(df2.Date[i])
X_test, y_test = np.array(X_test), np.array(y_test)

# Reshaping
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1],1))
```

```python
X_test.shape
```

```python
#Building the REAL Model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import GRU
from tensorflow import keras
```

```python
#GRU model with 128 units and 4 layers
model = Sequential()
model.add(GRU(units = 128,activation = 'relu', return_sequences = True, input_shape = (X_train.shape[1], 1)))
model.add(GRU(units = 128,activation = 'relu', return_sequences = True,dropout=0.2))
model.add(GRU(units = 128,activation = 'relu', return_sequences = True,dropout=0.2))
model.add(GRU(units = 128,activation = 'relu',dropout=0.2))

# Adding the output layer
model.add(Dense(units = 1))
```

```python
model.summary()
```

```python
# Compiling the RNN
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics='accuracy')

# Fitting the RNN to the Training set
history = model.fit(X_train, y_train,validation_data=(X_test,y_test), epochs = 10, batch_size = 64)
```

```python
# save the model to disk
model.save("modelmsft.h5")
```

```python
plt.figure(figsize=(8,6))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'validation loss'])
plt.show()
```

```python
#Predicting the test data
model = keras.models.load_model('modelmsft.h5')
prediction = model.predict(X_test)
```
